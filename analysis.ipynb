```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("/Users/luogan/Code/sglang/test_results/RR_router_test_20250731_090735.csv")
df2 = pd.read_csv("/Users/luogan/Code/sglang/test_results/RD_router_test_20250731_090823.csv")
df3 = pd.read_csv("/Users/luogan/Code/sglang/test_results/MU_router_test_20250731_090859.csv")
df4 = pd.read_csv("/Users/luogan/Code/sglang/test_results/MUR_router_test_20250731_090557.csv")
```

```python
r = "DECODE_THROUGHPUT"
df1 = df1.sort_values(by=r, ascending=True)
df2 = df2.sort_values(by=r, ascending=True)
df3 = df3.sort_values(by=r, ascending=True)
df4 = df4.sort_values(by=r, ascending=True)
plt.plot(df1["finish_time"], df1[r], label="round robin")
plt.plot(df2["finish_time"], df2[r], label="random")
plt.plot(df3["finish_time"], df3[r], label="marginal(ours)")
plt.plot(df4["finish_time"], df4[r], label="marginal record")
plt.xlabel("finish time")
plt.ylabel("decode throughput")
plt.grid()
plt.title("uniform, n_node=3")
plt.legend()
plt.savefig("decode_throughput_uniform.png")
plt.show()

r = "AVG_LATENCY"
df1 = df1.sort_values(by=r, ascending=True)
df2 = df2.sort_values(by=r, ascending=True)
df3 = df3.sort_values(by=r, ascending=True)
df4 = df4.sort_values(by=r, ascending=True)
plt.plot(df1["finish_time"], df1[r], label="round robin")
plt.plot(df2["finish_time"], df2[r], label="random")
plt.plot(df3["finish_time"], df3[r], label="marginal(ours)")
plt.plot(df4["finish_time"], df4[r], label="marginal record")
plt.xlabel("finish time")
plt.ylabel("average latency")
plt.yscale("log")
plt.grid()
plt.title("uniform, n_node=3")
plt.legend()
plt.savefig("average_latency_uniform.png")
plt.show()

r = "PREFILL_THROUGHPUT"
df1 = df1.sort_values(by=r, ascending=True)
df2 = df2.sort_values(by=r, ascending=True)
df3 = df3.sort_values(by=r, ascending=True)
df4 = df4.sort_values(by=r, ascending=True)
plt.plot(df1["finish_time"], df1[r], label="round robin")
plt.plot(df2["finish_time"], df2[r], label="random")
plt.plot(df3["finish_time"], df3[r], label="marginal(ours)")
plt.plot(df4["finish_time"], df4[r], label="marginal record")
plt.xlabel("finish time")
plt.ylabel("prefill throughput")
plt.grid()
plt.title("uniform, n_node=3")
plt.savefig("prefill_throughput_uniform.png")
plt.legend()
plt.show()
```
